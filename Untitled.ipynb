```python
import numpy as np
import pandas as pd
```

```python
activity = pd.read_csv("Final dataset files/dailyActivity_merged.csv")
heart = pd.read_csv("Final dataset files/heartrate_seconds_merged.csv")
sleep = pd.read_csv("Final dataset files/minuteSleep_merged.csv")
```

```python
activity.rename(columns={
    'ActivityDate': 'Date',
    'TotalSteps': 'Steps'
}, inplace=True)
activity['Date'] = pd.to_datetime(activity['Date'])
activity = activity[['Id', 'Date', 'Steps', 'Calories']]
```

```python
heart['Time'] = pd.to_datetime(heart['Time'])
heart['Date'] = heart['Time'].dt.date
heart['Date'] = pd.to_datetime(heart['Date'])
heart_daily = heart.groupby(['Id', 'Date'])['Value'].mean().reset_index()
heart_daily.rename(columns={'Value': 'HeartRate'}, inplace=True)
```

```python
sleep['date'] = pd.to_datetime(sleep['date'])
sleep['Date'] = sleep['date'].dt.date
sleep['Date'] = pd.to_datetime(sleep['Date'])
sleep_daily = sleep.groupby(['Id', 'Date'])['value'].sum().reset_index()
sleep_daily.rename(columns={'value': 'SleepMinutes'}, inplace=True)
```

```python
df = activity.merge(heart_daily, on=['Id', 'Date'], how='inner')
df = df.merge(sleep_daily, on=['Id', 'Date'], how='inner')
```

```python
df = df.sort_values(by='Date')
df = df.ffill()
```

```python
df['HR_Rolling_Mean'] = df['HeartRate'].rolling(window=7).mean()
df['HR_Rolling_Std'] = df['HeartRate'].rolling(window=7).std()
df['Z_score'] = (df['HeartRate'] - df['HR_Rolling_Mean']) / df['HR_Rolling_Std']
```

```python
def risk(z):
    if abs(z) < 1:
        return 20
    elif abs(z) < 2:
        return 50
    else:
        return 80
df['Risk_Score'] = df['Z_score'].apply(risk)
```

```python
df['Alert'] = df['Risk_Score'].apply(lambda x: "Yes" if x >= 80 else "No")
```

```python
df.to_csv("final_health_data.csv", index=False)
```
